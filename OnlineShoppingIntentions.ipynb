```python
from sklearn.model_selection import train_test_split
y=df_clean["Revenue"]
X_with=df_clean.drop(columns=["Revenue"])
```

```python
Xw_train,Xw_test,y_train,y_test=train_test_split(X_with,y,test_size=0.2,random_state=40,stratify=y)
```

```python
y_train.value_counts(normalize=True)
y_test.value_counts(normalize=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_with=DecisionTreeClassifier(random_state=40,class_weight="balanced")
dt_with.fit(Xw_train,y_train)
```

```python
y_pred=dt_with.predict(Xw_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score
print (accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(roc_auc_score(y_test,y_pred))
```

The dataset has an imbalanced class distribution.
The number of negative samples is higher than the number of positive samples.
Therefore, relying solely on the accuracy metric may be misleading, so the ROC-AUC metric was also evaluated.

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_depth': [3,5,7,10,None],
    "min_samples_leaf":[2,5,10],
    "criterion":["gini","entropy"]
    }
```

```python
grid=GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring="roc_auc",
    n_jobs=-1)
```

```python
grid.fit(Xw_train, y_train)
```

```python
best_dt = grid.best_estimator_

y_pred = grid.predict(Xw_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,roc_auc_score

print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(roc_auc_score(y_test, y_pred))
```

Hyperparameter tuning improved the Decision Tree model performance significantly.

Test accuracy increased from 0.86 to 0.90, and recall for the positive class improved from 0.51 to 0.58

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
plot_tree(
    grid.best_estimator_,
    feature_names=Xw_train.columns,
    class_names=["No Purchase", "Purchase"],
    filled=True,
    max_depth=3
    )
```

```python
#without PageValues
```

```python
from sklearn.model_selection import train_test_split

y = df_clean["Revenue"]
X_without = df_clean.drop(["Revenue", "PageValues"], axis=1)
```

The PageValues variable was excluded to prevent data leakage and ensure a reliable model evaluation.

```python
Xwo_train,Xwo_test,y_train,y_test=train_test_split(X_without,y,test_size=0.2,random_state=40,stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_without=DecisionTreeClassifier(random_state=40,class_weight="balanced")
dt_without.fit(Xwo_train,y_train)
```

```python
y_pred_without=dt_without.predict(Xwo_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score

print (accuracy_score(y_test,y_pred_without))
print(confusion_matrix(y_test,y_pred_without))
print(classification_report(y_test,y_pred_without))
print(roc_auc_score(y_test,y_pred_without))
```

To improve generalization performance and obtain a more robust model, a Random Forest classifier will be trained using the same train-test split.

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=400,
    max_depth=None,
    min_samples_leaf=5,
    class_weight="balanced",
    random_state=40,
    n_jobs=-1
)

rf.fit(Xwo_train, y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report, roc_auc_score

y_pred_rf = rf.predict(Xwo_test)
y_proba_rf = rf.predict_proba(Xwo_test)[:,1]

print(accuracy_score(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
print( roc_auc_score(y_test, y_proba_rf))
```

```python
importances = rf.feature_importances_
feature_names = Xwo_train.columns

feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(15), palette='viridis')
plt.title('Random Forest - Top 15 Feature Importances (Without PageValues)', fontsize=15)
plt.xlabel('Relative Importance Score')
plt.ylabel('Features')
plt.tight_layout()
```

In this project, customer purchasing behavior was predicted using Decision Tree and Random Forest models.

The dataset contained an imbalanced class distribution, therefore evaluation was not limited to accuracy. Metrics such as precision, recall, and ROC-AUC were also considered to better assess model performance.

The Random Forest model achieved more balanced and stable results compared to the Decision Tree model. This suggests that ensemble methods can improve generalization performance and reduce overfitting.

Additionally, the PageValues feature was removed to prevent overly optimistic results and ensure a fair evaluation.
